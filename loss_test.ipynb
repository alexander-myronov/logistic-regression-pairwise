```python
import pandas as pd
import numpy as np
import itertools
from matplotlib import pyplot as plt
```

```python
loss = pd.read_csv('data/cross_loss_tests.csv')
```

```python
loss.shape
```

```python
loss.loc[:, 'f1_on_D1_loss'] =  loss.f1_on_D1_labeled_loss + loss.f1_on_D1_links_loss + \
    loss.f1_on_D1_unlabeled_loss + loss.f1_on_D1_norm_loss
loss.loc[:, 'f1_on_D2_loss'] =  loss.f1_on_D2_labeled_loss + loss.f1_on_D2_links_loss + \
    loss.f1_on_D2_unlabeled_loss + loss.f1_on_D2_norm_loss
loss.loc[:, 'f2_on_D1_loss'] =  loss.f2_on_D1_labeled_loss + loss.f2_on_D1_links_loss + \
    loss.f2_on_D1_unlabeled_loss + loss.f2_on_D1_norm_loss
loss.loc[:, 'f2_on_D2_loss'] =  loss.f2_on_D2_labeled_loss + loss.f2_on_D2_links_loss + \
    loss.f2_on_D2_unlabeled_loss + loss.f2_on_D2_norm_loss
    
```

```python
loss.loc[:, 'labeled_fail'] = \
    (loss.f1_on_D1_labeled_loss >= loss.f2_on_D1_labeled_loss) | (loss.f2_on_D2_labeled_loss >= loss.f1_on_D2_labeled_loss)
    
loss.loc[:, 'link_fail'] = \
    (loss.f1_on_D1_links_loss >= loss.f2_on_D1_links_loss) | (loss.f2_on_D2_links_loss >= loss.f1_on_D2_links_loss)
loss.loc[:, 'unlabeled_fail'] = \
    (loss.f1_on_D1_unlabeled_loss >= loss.f2_on_D1_unlabeled_loss) | \
    (loss.f2_on_D2_unlabeled_loss >= loss.f1_on_D2_unlabeled_loss)
```

```python
loss.loc[:, 'loss_fail'] = \
    (loss.f1_on_D1_loss >= loss.f2_on_D1_loss) | (loss.f2_on_D2_loss >= loss.f1_on_D2_loss)
```

```python
loss.labeled_fail.mean(), loss.link_fail.mean(), loss.unlabeled_fail.mean()
```

```python
loss.groupby(['std','percent_labels', 'alpha']).agg({'labeled_fail': 'mean'}).unstack('std')
```

```python
loss.groupby(['std','percent_links', 'beta']).agg({'link_fail': 'mean'}).unstack('std')
```

```python
loss.groupby(['std','percent_unlabeled', 'delta']).agg({'unlabeled_fail': 'mean'}).unstack('std')
```

```python
loss.groupby(['std','percent_labels']).agg({'loss_fail': 'mean'}).unstack('std')
```
